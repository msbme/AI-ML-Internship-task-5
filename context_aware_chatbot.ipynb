```python
# ==================================================================================
# CONTEXT-AWARE RAG CHATBOT (LangChain) — COMPLETE Single Colab Cell (FIXED)
# Objective: Build a conversational chatbot with memory + retrieval over a custom
# knowledge base, deployed via Streamlit (tunneled with ngrok from Colab).
# NOTE: transformers is pinned to a stable version that supports the
# "text2text-generation" pipeline task (some newer transformers builds removed it).
# ==================================================================================

# ----------------------------------------------------------------------------------
# 1. INSTALL DEPENDENCIES (transformers pinned for pipeline compatibility)
# ----------------------------------------------------------------------------------
!pip install -q -U langchain-community langchain-core langchain-text-splitters \
                   langchain-huggingface sentence-transformers faiss-cpu \
                   "transformers==4.44.2" "accelerate==0.34.2" \
                   streamlit pyngrok matplotlib

import os
import re
import textwrap
import matplotlib.pyplot as plt

from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_community.llms import HuggingFacePipeline
from transformers import pipeline

print("✅ Libraries imported successfully.")

# ----------------------------------------------------------------------------------
# 2. PROBLEM STATEMENT & OBJECTIVE
# ----------------------------------------------------------------------------------
print(textwrap.dedent("""
PROBLEM STATEMENT:
Standard LLMs cannot answer questions about private/custom knowledge bases and
forget prior turns in a conversation. This project builds a Retrieval-Augmented
Generation (RAG) chatbot that:
  1. Loads and chunks a custom text knowledge base.
  2. Embeds chunks into a vector store for semantic retrieval.
  3. Maintains conversational memory across turns (implemented manually).
  4. Deploys an interactive chat UI via Streamlit.
"""))

# ----------------------------------------------------------------------------------
# 3. DATASET LOADING & PREPROCESSING
#    (Auto-prompts for upload if knowledge_base.txt is missing)
# ----------------------------------------------------------------------------------
DATA_PATH = "/content/knowledge_base.txt"

if not os.path.exists(DATA_PATH):
    print("⚠️  knowledge_base.txt not found in /content/. Please upload it now:")
    from google.colab import files
    uploaded = files.upload()

    if DATA_PATH.split("/")[-1] not in uploaded and len(uploaded) > 0:
        uploaded_filename = list(uploaded.keys())[0]
        os.rename(f"/content/{uploaded_filename}", DATA_PATH)
        print(f"📌 Renamed '{uploaded_filename}' to 'knowledge_base.txt'")

if not os.path.exists(DATA_PATH):
    raise FileNotFoundError(f"❌ Still could not find {DATA_PATH}. Please re-run the cell and upload the file.")

with open(DATA_PATH, "r", encoding="utf-8") as f:
    raw_text = f.read()

raw_text = re.sub(r"\s+", " ", raw_text).strip()
print(f"📄 Loaded knowledge base — {len(raw_text)} characters.")

doc = Document(page_content=raw_text, metadata={"source": "knowledge_base.txt"})

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=80,
    separators=["\n\n", "\n", ". ", " ", ""]
)
chunks = text_splitter.split_documents([doc])
print(f"✂️  Split into {len(chunks)} chunks.")

# ----------------------------------------------------------------------------------
# 4. BUILD VECTORIZED DOCUMENT STORE (EMBEDDINGS + FAISS)
# ----------------------------------------------------------------------------------
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

vectorstore = FAISS.from_documents(chunks, embedding_model)
VECTORSTORE_PATH = "/content/faiss_index"
vectorstore.save_local(VECTORSTORE_PATH)
print(f"🗂️  FAISS vector store built and saved to {VECTORSTORE_PATH}")

# ----------------------------------------------------------------------------------
# 5. MODEL DEVELOPMENT — LLM SETUP (local, no API key required)
# ----------------------------------------------------------------------------------
gen_pipeline = pipeline(
    "text2text-generation",
    model="google/flan-t5-base",
    max_length=512,
    temperature=0.3
)
llm = HuggingFacePipeline(pipeline=gen_pipeline)
print("🤖 LLM (google/flan-t5-base) loaded.")

# ----------------------------------------------------------------------------------
# 6. MANUAL CONVERSATIONAL MEMORY + RETRIEVAL LOGIC
#    (Avoids langchain.memory / langchain.chains entirely for stability)
# ----------------------------------------------------------------------------------
chat_history = []  # list of (question, answer) tuples — our "memory"

def build_prompt(question, context, history, max_turns=3):
    history_text = ""
    for q, a in history[-max_turns:]:
        history_text += f"Previous Question: {q}\nPrevious Answer: {a}\n"

    prompt = f"""Answer the question using ONLY the context below. If the answer
isn't in the context, say "I don't have enough information to answer that."

Context:
{context}

Conversation so far:
{history_text}

New Question: {question}
Answer:"""
    return prompt

def ask(question, k=3, verbose=False):
    retrieved_docs = vectorstore.similarity_search(question, k=k)
    context = "\n\n".join([d.page_content for d in retrieved_docs])

    prompt = build_prompt(question, context, chat_history)
    answer = llm.invoke(prompt).strip()

    chat_history.append((question, answer))

    if verbose:
        print(f"\n[Retrieved {len(retrieved_docs)} chunks]")
        for i, d in enumerate(retrieved_docs):
            print(f"  source {i+1}: {d.page_content[:80]}...")

    return answer, retrieved_docs

print("🔗 Manual RAG chain with memory ready.")

# ----------------------------------------------------------------------------------
# 7. EVALUATION — Quick functional test with sample queries
# ----------------------------------------------------------------------------------
sample_questions = [
    "What is this document about?",
    "Can you give me more detail on that?",
]

print("\n========== SAMPLE CONVERSATION TEST ==========")
similarity_scores_log = []
for q in sample_questions:
    answer, _ = ask(q, verbose=True)
    print(f"\nQ: {q}")
    print(f"A: {answer}")

    scored_docs = vectorstore.similarity_search_with_score(q, k=3)
    similarity_scores_log.append([score for _, score in scored_docs])

# ----------------------------------------------------------------------------------
# 8. VISUALIZATIONS
# ----------------------------------------------------------------------------------
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

chunk_lengths = [len(c.page_content) for c in chunks]
axes[0].hist(chunk_lengths, bins=15, color="#4C72B0", edgecolor="black")
axes[0].set_title("Distribution of Chunk Lengths")
axes[0].set_xlabel("Characters per chunk")
axes[0].set_ylabel("Frequency")

axes[1].bar(range(1, len(similarity_scores_log[-1]) + 1), similarity_scores_log[-1], color="#DD8452")
axes[1].set_title("Top-k Retrieval Distance (last query)")
axes[1].set_xlabel("Retrieved chunk rank")
axes[1].set_ylabel("L2 distance (lower = better match)")

plt.tight_layout()
plt.savefig("/content/eval_visualizations.png")
plt.show()
print("📊 Visualizations saved to /content/eval_visualizations.png")

# ----------------------------------------------------------------------------------
# 9. STREAMLIT APP — write app.py to disk
# ----------------------------------------------------------------------------------
streamlit_code = '''
import streamlit as st
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_community.llms import HuggingFacePipeline
from transformers import pipeline

st.set_page_config(page_title="Context-Aware RAG Chatbot", page_icon="💬")
st.title("💬 Context-Aware Knowledge Base Chatbot")
st.caption("Powered by LangChain + FAISS + flan-t5-base — remembers conversation context.")

VECTORSTORE_PATH = "/content/faiss_index"

@st.cache_resource
def load_resources():
    embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
    vectorstore = FAISS.load_local(VECTORSTORE_PATH, embedding_model, allow_dangerous_deserialization=True)
    gen_pipeline = pipeline("text2text-generation", model="google/flan-t5-base", max_length=512, temperature=0.3)
    llm = HuggingFacePipeline(pipeline=gen_pipeline)
    return vectorstore, llm

vectorstore, llm = load_resources()

if "chat_history" not in st.session_state:
    st.session_state.chat_history = []
if "messages" not in st.session_state:
    st.session_state.messages = []

def build_prompt(question, context, history, max_turns=3):
    history_text = ""
    for q, a in history[-max_turns:]:
        history_text += f"Previous Question: {q}\\nPrevious Answer: {a}\\n"
    return f"""Answer the question using ONLY the context below. If the answer
isn't in the context, say "I don't have enough information to answer that."

Context:
{context}

Conversation so far:
{history_text}

New Question: {question}
Answer:"""

def ask(question, k=3):
    retrieved_docs = vectorstore.similarity_search(question, k=k)
    context = "\\n\\n".join([d.page_content for d in retrieved_docs])
    prompt = build_prompt(question, context, st.session_state.chat_history)
    answer = llm.invoke(prompt).strip()
    st.session_state.chat_history.append((question, answer))
    return answer, retrieved_docs

for msg in st.session_state.messages:
    with st.chat_message(msg["role"]):
        st.markdown(msg["content"])

user_input = st.chat_input("Ask something about the knowledge base...")

if user_input:
    st.session_state.messages.append({"role": "user", "content": user_input})
    with st.chat_message("user"):
        st.markdown(user_input)

    with st.spinner("Thinking..."):
        answer, sources = ask(user_input)

    with st.chat_message("assistant"):
        st.markdown(answer)
        if sources:
            with st.expander("📚 Sources used"):
                for i, d in enumerate(sources):
                    st.markdown(f"**Source {i+1}:** {d.page_content[:300]}...")

    st.session_state.messages.append({"role": "assistant", "content": answer})

with st.sidebar:
    st.header("About")
    st.write("This chatbot retrieves answers from your custom knowledge_base.txt using RAG, and remembers the conversation history.")
    if st.button("🗑️ Clear conversation"):
        st.session_state.messages = []
        st.session_state.chat_history = []
        st.rerun()
'''

with open("/content/app.py", "w", encoding="utf-8") as f:
    f.write(streamlit_code)

print("📝 Streamlit app written to /content/app.py")

# ----------------------------------------------------------------------------------
# 10. DEPLOY VIA STREAMLIT + NGROK
# ----------------------------------------------------------------------------------
from pyngrok import ngrok

NGROK_AUTH_TOKEN = "YOUR_NGROK_AUTH_TOKEN"

if NGROK_AUTH_TOKEN != "YOUR_NGROK_AUTH_TOKEN":
    ngrok.set_auth_token(NGROK_AUTH_TOKEN)
    ngrok.kill()
    public_url = ngrok.connect(8501)
    print(f"🌐 Your chatbot is live at: {public_url}")
    get_ipython().system_raw("streamlit run /content/app.py --server.port 8501 &")
else:
    print("⚠️  Set NGROK_AUTH_TOKEN above to deploy publicly. "
          "Until then, the app.py file is ready at /content/app.py.")

# ----------------------------------------------------------------------------------
# 11. FINAL SUMMARY / INSIGHTS
# ----------------------------------------------------------------------------------
print(textwrap.dedent(f"""
========================== SUMMARY ==========================
- Knowledge base loaded from: {DATA_PATH}
- Total chunks indexed: {len(chunks)}
- Embedding model: sentence-transformers/all-MiniLM-L6-v2
- Vector store: FAISS (saved at {VECTORSTORE_PATH})
- LLM: google/flan-t5-base (local, no API key needed)
- Memory: Manual conversation buffer (list of Q&A tuples, last 3 turns used as context)
- Deployment: Streamlit app at /content/app.py, tunneled via ngrok
===============================================================
"""))
```
